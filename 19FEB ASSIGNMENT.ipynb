Q1. What is Flask Framework? What are the advantages of Flask Framework?

ANS-Flask is a micro web framework written in Python. It is classified as a microframework because it does not require particular tools or libraries. It is based on the Werkzeug WSGI toolkit and Jinja2 template engine.

Advantages of using Flask framework are:


There is a built-in development server and a fast debugger provided.


lightweight

Support Secure cookies.


Templating using Jinja2.


Request dispatching using REST.


Support for unit testing is built-in.

Q3. What is App routing in Flask? Why do we use app routes?

ANS-App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. 

App routes are used in Python to map specific URLs to functions that handle those URLs. This allows developers to create more complex and dynamic web applications.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

ANS-The Flask url_for() function is used to generate a URL to a specific function. It takes the name of the function as its first argument, and any number of keyword arguments, each corresponding to a variable part of the URL rule.
