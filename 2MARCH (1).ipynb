Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

ANS-A scatter plot is a type of data visualization that displays the relationship between two variables. Each data point in the plot represents the value of one variable against the value of another.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
```

```python
plt.scatter(x,y)
plt.title('This is title')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

```python
import numpy as np
#line1: 
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
#line2: 
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
#line3: 
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
#line4: 
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

ANS-
The subplot() function in Matplotlib is used to create a grid of plots within a single figure. It allows you to create multiple plots (subplots) arranged in rows and columns, making it easier to compare different datasets or visualizations.

```python
plt.figure(figsize=(10, 8))
```

```python
# Creating subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, color='blue')
plt.title('Line 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, color='orange')
plt.title('Line 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, color='green')
plt.title('Line 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, color='red')
plt.title('Line 4')



# Show plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

ANS-A bar plot is a graphical representation of categorical data with rectangular bars whose lengths are proportional to the values they represent. It's typically used to compare the values of different categories or to track changes over time within a single category.

```python
# Creating a vertical bar plot
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.title('Profit of Companies')

plt.show()

# Creating a horizontal bar plot
plt.figure(figsize=(8, 5))
plt.barh(company, profit, color='lightgreen')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.title('Profit of Companies')

plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

ANS_A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a visual summary of the central tendency, dispersion, and skewness (asymmetry) of the data.

Box plots are used to visually compare distributions of different datasets or to identify the presence of outliers. They are particularly useful when dealing with large datasets or multiple datasets.

```python
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Values')
plt.show()
```
