```python
from bokeh.plotting import figure , output_file , show
from bokeh.sampledata.iris import flowers

output_file('test.html')

p = figure(title='test flowers')
p.xaxis.axis_label = 'sepal_length'
p.yaxis.axis_label = 'sepal_width'
p.circle(flowers['sepal_length'], flowers['sepal_width'])
show(p)
```

```python
from bokeh.plotting import figure , output_file , show
from bokeh.sampledata.iris import flowers

output_file('line.html')

p = figure(title='line graph')
p.xaxis.axis_label = 'x axis'
p.yaxis.axis_label = 'y axis'
p.line(flowers['sepal_length'], flowers['sepal_width'])
show(p)
```

```python
x = [2,3,4,5,6]
y = [4,5,6,7,8]
output_file('line.html')
p = figure(title= 'line plot')
p.line(x,y)
show(p)
```

```python
x = [2,3,4,5,6]
y = [4,5,6,7,8]
output_file('line.html')
p = figure(title= 'line plot')
p.scatter(x,y,fill_color = 'red',legend_label = 'red points',size = 20)
show(p)
```
