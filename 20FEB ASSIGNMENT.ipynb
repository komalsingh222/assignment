Q1. Explain GET and POST methods.

ANS-The GET and POST methods are two of the most common HTTP methods used in Flask.

The GET method is used to retrieve data from the server. It is typically used for requests that do not require any changes to be made to the server's data. 

The POST method is used to submit data to the server. It is typically used for requests that require changes to be made to the server's data. 

Q2. Why is request used in Flask?

ANS-The request object in Flask is used to access the data sent from the client to the server. This data can include form data, cookies, headers, and more. The request object is available in all Flask views, and can be used to process the data sent by the client.

Q3. Why is redirect() used in Flask?

ANS-The redirect() function in Flask is used to send the user to a different URL.

Q4. What are templates in Flask? Why is the render_template() function used?

ANS-In Flask, templates are HTML files that contain static data as well as placeholders for dynamic data. A template is rendered with specific data to produce a final document. Flask uses the Jinja template library to render templates.

The render_template() function is used to render a template file and return the resulting HTML. This function takes two arguments: the name of the template file and a dictionary of variables that will be passed to the template.
